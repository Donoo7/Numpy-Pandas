```python
import seaborn as sns 
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x ='total_bill', y ='tip', data=tips)
plt.title('Total Bill vs. Tip Amount')
plt.show()
```

```python
sns.histplot(tips['total_bill'], kde=True)
plt.title('Distribution of total bill amount')
plt.show()
plt.savefig('total_bill_distribution.png')
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('total bill amount by day')
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip',data=tips,hue='sex')
plt.title('total bill vs. tip amount by grnder')
plt.show()
```

```python
corr = tips.corr()
plt.figure(figsize = (8, 6))
sns.heatmap(corr, annot = True, cmap = 'BuPu', fmt = '.3f')
plt.title('correlation heatmap pf numerical variables in tips datasets')
plt.show()
```
